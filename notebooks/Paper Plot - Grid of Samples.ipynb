```python
def set_seeds(seed):
    torch.use_deterministic_algorithms(True)
    random.seed(seed)
    np.random.seed(seed)
    torch.manual_seed(seed)
```

```python
sampling_data = sampler.load_sampling_data()
```

```python
weights_file = 'checkpoints/experiment59-cs-fix-and-context-scaling_16.pth'
```

```python
imgs_max = []

for i in tqdm.tqdm(range(50)):
    img, _ = sampler.sample(
        weights_file=weights_file,
        sampling_data=sampling_data,
        context=(50, 50, 10, 10, 1, -1),
        method='ddpm',
    )
    imgs_max.append(img)
```

```python
imgs_min = []

for i in tqdm.tqdm(range(50)):
    img, _ = sampler.sample(
        weights_file=weights_file,
        sampling_data=sampling_data,
        context=(0, 0, 0, 0, 1, -1),
        method='ddpm',
    )
    imgs_min.append(img)
```

```python
def plot_grid(imgs, title):
    fig, axes = plt.subplots(8, 6)
    axes = axes.flatten()
    for i, ax in enumerate(axes):
        L = np.abs(imgs[i][0]).max() * .7
        ax.imshow(imgs[i][0], origin='lower', cmap='coolwarm', vmin=-L, vmax=L)
        ax.set_xticks([])
        ax.set_yticks([])
    plt.subplots_adjust(wspace=0.0, hspace=.0)
    fig.suptitle(title)
    return fig
```

```python
fig = plot_grid(imgs_min, 'Grid of Samples: Solar Minimum Case\n(Magnetic Potential at $r=1~R_\\odot$)')
fig.savefig('plots/paper/50samples_min.png', dpi=300)
```

```python
fig = plot_grid(imgs_max, 'Grid of Samples: Solar Maximum Case\n(Magnetic Potential at $r=1~R_\\odot$)')
fig.savefig('plots/paper/50samples_max.png', dpi=300)
```
